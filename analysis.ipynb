```python
import json
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
import os
import json

folder_path = './gas_snapshots' 

# List your JSON file names; change these if your filenames differ
file_names = ["preSorted.json", "runtimeSort.json", "runtimeSortTransient.json"]

# Create a dictionary to hold each method's data
methods = {}
for file_name in file_names:
    full_path = os.path.join(folder_path, file_name)
    with open(full_path, 'r') as f:
        data = json.load(f)
    # Use the file name (without .json) as the method name
    method_name = file_name.rsplit('.', 1)[0]
    methods[method_name] = data

print("Loaded methods:", list(methods.keys()))
```

```python
def plot_graph(numAssets):
    plt.figure(figsize=(8, 5))
    for method_name, data in methods.items():
        numSupplied_values = []
        gas_used_values = []
        # Keys are expected to follow the format: numAssets_{numAssets}_numSupplied_{numSupplied}
        for key, value in data.items():
            parts = key.split("_")
            asset_count = int(parts[1])
            supplied_count = int(parts[3])
            if asset_count == numAssets:
                numSupplied_values.append(supplied_count)
                gas_used_values.append(int(value))
        if numSupplied_values:  # Only plot if data exists for this numAssets
            # Sort by numSupplied values to ensure a proper line plot
            numSupplied_values, gas_used_values = zip(*sorted(zip(numSupplied_values, gas_used_values)))
            plt.plot(numSupplied_values, gas_used_values, marker='o', label=method_name)
    plt.xlabel("numSupplied")
    plt.ylabel("Gas Used")
    plt.title(f"Gas Usage Comparison for numAssets = {numAssets}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Adjust the slider's max value as needed based on your data
numAssets_slider = widgets.IntSlider(value=1, min=1, max=128, step=1, description='numAssets')
widgets.interactive(plot_graph, numAssets=numAssets_slider)
```
